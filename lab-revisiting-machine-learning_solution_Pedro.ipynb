```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('learningSet.csv')
df
```

```python
df.isna().sum()
```

```python
drop_list=[]
drop_list = ['ZIP','OSOURCE']
```

```python
nulls_perc = pd.DataFrame(df.isna().sum()/df.shape[0])
list1 = list(nulls_perc[nulls_perc[0]>0.85].index)

drop_list1 = drop_list + list1

df = df.drop(drop_list1, axis = 1)
```

```python
df.shape
```

```python
df['GENDER'].value_counts()

def clean_gender(x):
    if 'M' in x:
        return 'M'
    elif 'F' in x:
        return 'F'
    else:
        return 'other'
    
df['GENDER'] = df['GENDER'].apply(clean_gender)
df['GENDER'].value_counts(dropna=False)
```
